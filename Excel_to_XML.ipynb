```python
import pandas as pd

df = pd.read_excel (r'14_Topik_14.xlsx')
print (df)
```

```python
import base64

topik = ["<?xml version=\"1.0\" encoding=\"UTF-8\"?>"]
topik.append("<quiz>")
topik.append("<!-- question: 0  -->")
topik.append("<question type=\"category\">")
topik.append("<category>")
topik.append("<text>$course$/UAS/MultipleChoice/Topik 14</text>")
topik.append("</category>")
topik.append("</question>")    

for index, row in df.iterrows():
    topik.append("<question type=\"multichoice\">")
    topik.append("<name>")
    topik.append("<text>")
    topik.append(str(row["No"]))
    topik.append("</text>")
    topik.append("</name>")    
    topik.append("<questiontext format=\"html\">")
    topik.append("<text>")
    topik.append("<![CDATA[<p> "+str(row["Soal"])+"<br></p>]]>")    
    topik.append("</text>")        
    topik.append("</questiontext>")
    topik.append("<generalfeedback format=\"html\">")
    topik.append("<text></text>")
    topik.append("</generalfeedback>")
    topik.append("<defaultgrade>1.0000000</defaultgrade>")
    topik.append("<penalty>0.3333333</penalty>")
    topik.append("<hidden>0</hidden>")
    topik.append("<idnumber></idnumber>")
    topik.append("<single>true</single>")
    topik.append("<shuffleanswers>true</shuffleanswers>")
    topik.append("<answernumbering>abc</answernumbering>")
    if str(row["Soal Gambar"]) != "nan":
        topik.append("<image>"+str(row["Soal Gambar"])+"</image>")
        with open(row["Soal Gambar"], "rb") as imageFile:
            imgstr = base64.b64encode(imageFile.read())
            image = imgstr.decode("utf-8")
            topik.append("<image_base64>")
            topik.append(str(image))
            topik.append("</image_base64>")
    topik.append("<correctfeedback format=\"html\"><text>Jawaban anda benar.</text></correctfeedback>")
    topik.append("<partiallycorrectfeedback format=\"html\"><text>Jawaban anda separuh benar.</text></partiallycorrectfeedback>")
    topik.append("<incorrectfeedback format=\"html\"><text>Jawaban anda tidak benar.</text></incorrectfeedback>")
    topik.append("<shownumcorrect />")
    kunci=row["Kunci Jawaban"]
    if (row["Opsi A"]==row["Kunci Jawaban"]) or (kunci == "A") or (kunci == "a"):
        topik.append("<answer fraction=\"100\" format=\"html\">")
    else:
        topik.append("<answer fraction=\"0\" format=\"html\">")
    topik.append("<text>")
    topik.append("<![CDATA[<p>"+str(row["Opsi A"])+"<br></p>]]>")
    topik.append("</text>")
    topik.append("<feedback format=\"html\"><text></text></feedback>")
    topik.append("</answer>")
    if (row["Opsi B"]==row["Kunci Jawaban"]) or (kunci == "B") or (kunci == "b"):
        topik.append("<answer fraction=\"100\" format=\"html\">")
    else:
        topik.append("<answer fraction=\"0\" format=\"html\">")
    topik.append("<text>")
    topik.append("<![CDATA[<p>"+str(row["Opsi B"])+"<br></p>]]>")
    topik.append("</text>")
    topik.append("<feedback format=\"html\"><text></text></feedback>")
    topik.append("</answer>")
    if (row["Opsi C"]==row["Kunci Jawaban"]) or (kunci == "C") or (kunci == "c"):
        topik.append("<answer fraction=\"100\" format=\"html\">")
    else:
        topik.append("<answer fraction=\"0\" format=\"html\">")
    topik.append("<text>")
    topik.append("<![CDATA[<p>"+str(row["Opsi C"])+"<br></p>]]>")
    topik.append("</text>")
    topik.append("<feedback format=\"html\"><text></text></feedback>")
    topik.append("</answer>")
    if (row["Opsi D"]==row["Kunci Jawaban"]) or (kunci == "D") or (kunci == "d"):
        topik.append("<answer fraction=\"100\" format=\"html\">")
    else:
        topik.append("<answer fraction=\"0\" format=\"html\">")
    topik.append("<text>")
    topik.append("<![CDATA[<p>"+str(row["Opsi D"])+"<br></p>]]>")
    topik.append("</text>")
    topik.append("<feedback format=\"html\"><text></text></feedback>")
    topik.append("</answer>")
    
    topik.append("</question>")
topik.append("</quiz>")

res='\n'.join(topik)

print(res)

filename="14_Topik_14.xml"
with open(filename, 'w') as f:
     f.write(res)
```
